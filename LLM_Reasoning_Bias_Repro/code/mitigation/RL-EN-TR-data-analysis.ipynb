```python
import re

reasoning = data[0]["reasoning"]

# Regex to match "Step X: ..." and capture the step number and its content
pattern = r"(Step \d+:)(.*?)(?=Step \d+:|$)"

steps = re.findall(pattern, reasoning, re.DOTALL)

# Convert to a list of (step_header, step_content)
parsed_steps = [(header.strip(), content.strip()) for header, content in steps]

for header, content in parsed_steps:
    print(header)
    print(content)
    print("---")
```
